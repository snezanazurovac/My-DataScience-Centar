```python
import pandas as pd
df = pd.read_csv('../input/daycsv/day.csv')
dep_var = 'cnt'
cat_names = ['season', 'yr', 'mnth', 'holiday', 'weekday','workingday', 'weathersit']
cont_names = ['instant','season','yr','mnth','holiday','weekday','workingday','weathersit','temp','hum','windspeed']
procs = [Categorify, Normalize]
```

```python
df.head()
df.info()
```

```python
data = (TabularList.from_df(df,cont_names=cont_names, procs=procs)
                           .split_by_idx(valid_idx=range(300,500))
                           .label_from_df(cols=dep_var)
                           .databunch())
```

```python
data.show_batch()
```

```python
learn = tabular_learner(data, layers=[200,100], metrics=accuracy)
learn.fit(5, 1e-2)
#learn.save('mini_train')
```

```python
learn.show_results()
```
