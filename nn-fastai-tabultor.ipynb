```python
import pandas as pd
df = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
dep_var = 'quality'
#cat_names = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']
cont_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
procs = [FillMissing, Normalize]
```

```python
df.head()
df.info()
```

```python
data = (TabularList.from_df(df,cont_names=cont_names, procs=procs)
                           .split_by_idx(valid_idx=range(1280,1599))
                           .label_from_df(cols=dep_var)
                           .databunch())
```

```python
data.show_batch()
```

```python
learn = tabular_learner(data, layers=[200,100], metrics=accuracy)
learn.fit(5, 1e-2)
#learn.save('mini_train')
```

```python
learn.show_results()
```
