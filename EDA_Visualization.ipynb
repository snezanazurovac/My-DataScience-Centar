```python
fig = plt.gcf()
fig.set_size_inches(10,10)
pc = data[['math score','reading score','writing score']].corr(method ='pearson')

cols = ['math score','reading score','writing score']

ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 30})
```

```python
import os #provides functions for interacting with the operating system
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
fig = plt.gcf()
fig.set_size_inches(10,10)
data = pd.read_csv("data_students.csv")
# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(x="parental level of education", y="math score", hue="gender",split=True, inner="quartile",palette=None,
               data=data)
sns.despine(left=True)
```
